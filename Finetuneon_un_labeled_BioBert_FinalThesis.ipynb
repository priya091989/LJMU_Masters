```python
# Tokenization function
def tokenize_function(examples):
    # Join all contexts into one text chunk for MLM
    texts = []
    for ctx in examples["context"]:
        texts.append(" ".join(ctx["contexts"]))
    # Combine question + contexts for MLM
    combined_texts = [q + " " + c for q, c in zip(examples["question"], texts)]
    return tokenizer(combined_texts, truncation=True, max_length=256, padding="max_length")
```

```python
from transformers import AutoModelForMaskedLM, TrainingArguments, Trainer
from sklearn.metrics import accuracy_score, f1_score

model = AutoModelForMaskedLM.from_pretrained(
    MODEL_NAME)

def compute_metrics(pred):
    preds = pred.predictions.argmax(-1)
    return {
        "accuracy": accuracy_score(pred.label_ids, preds),
        "f1": f1_score(pred.label_ids, preds, average="weighted")
    }

# Training arguments
training_args = TrainingArguments(
    output_dir="./biobert_pubmedqa_mlm",
    overwrite_output_dir=True,
    num_train_epochs=8,
    per_device_train_batch_size=16,
    save_steps=100,
    save_total_limit=2,
    logging_steps=100,
    eval_strategy="no",
    learning_rate=5e-5,
    weight_decay=0.01,
    fp16=True,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,  # <-- important!
)

trainer.train()

model.save_pretrained("./biobert-pretrained")
tokenizer.save_pretrained("./biobert-pretrained")

# Summary:
#Let me know if you want a complete runnable script with this two-stage pipeline (MLM → QA classification using PubMedQA).
```

```python
#🧠 Step 2: Define preprocessing and label mapping

from transformers import AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("/content/biobert_pubmedqa_mlm/checkpoint-500")  # or the one used earlier

label_map = {"yes": 0, "no": 1, "maybe": 2}

def preprocess(example):
    question = example["question"]
    context = example["context"]
    answer = example["final_decision"]

    choices = ["yes", "no", "maybe"]
    inputs = tokenizer(
        [question] * 3,
        [context + " " + choice for choice in choices],
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=512,
    )
    label = label_map[answer]
    return inputs, label
```

```python
from transformers import AutoModel
import torch.nn as nn

class QAClassifier(nn.Module):
    def __init__(self, base_model_name):
        super(QAClassifier, self).__init__()
        self.encoder = AutoModel.from_pretrained(base_model_name)
        self.classifier = nn.Linear(self.encoder.config.hidden_size, 1)

    def forward(self, input_ids, attention_mask):
        outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]  # CLS token
        logits = self.classifier(cls_output)
        return logits.view(-1, 3)  # reshape to (batch_size, 3 choices)

# Load your MLM-trained model here
qa_model = QAClassifier("/content/biobert_pubmedqa_mlm/checkpoint-500")
# qa_model.load_state_dict(torch.load("/content/biobert_pubmedqa_mlm/checkpoint-500"))
qa_model.eval()
```

```python
def preprocess(example):
    question = example["question"]

    # Fix context to extract text from dict
    context = example["context"]
    if isinstance(context, dict):
        context = context.get("text", "")

    answer = example["final_decision"]

    choices = ["yes", "no", "maybe"]
    inputs = tokenizer(
        [question] * 3,
        [context + " " + choice for choice in choices],
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=512,
    )
    label = label_map[answer]
    return inputs, label
```

```python
# Evaluation loop
correct = 0
total = 0
qa_model.eval()

for example in labeled_pubmedqa:
    inputs, label = preprocess(example)
    with torch.no_grad():
        logits = qa_model(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"]
        )
        pred = torch.argmax(logits, dim=1).item()
    if pred == label:
        correct += 1
    total += 1

print(f"Accuracy: {correct / total:.2%}")
```

```python
for example in labeled_pubmedqa:
    inputs, label = preprocess(example)
    with torch.no_grad():
        logits = qa_model(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"]
        )
        pred_label_id = torch.argmax(logits, dim=1).item()
        pred_label = inv_label_map[pred_label_id]

    # Compare predicted label string with ground truth answer
    if pred_label == example["final_decision"]:
        correct += 1
    total += 1

print(f"Exact Match Accuracy: {correct / total:.2%}")
```

```python
import numpy as np
```

```python
all_logits = []
all_true = []
all_preds = []

with torch.no_grad():
    for example in labeled_pubmedqa:
        inputs, label = preprocess(example)  # label is int, e.g. 0,1,2

        logits = qa_model(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"]
        )  # shape: likely (3, num_classes)

        logits = logits.cpu()

        # Convert logits to probabilities
        probs = torch.softmax(logits, dim=1)  # shape (3, num_classes)

        all_logits.append(probs.numpy())
        all_true.append(label)

        # Predict choice: the choice with highest max probability over classes
        choice_scores = probs.max(dim=1).values  # max prob per choice
        pred_choice = torch.argmax(choice_scores).item()
        all_preds.append(pred_choice)
```

```python
from sklearn.metrics import average_precision_score
```

```python
# Calculate Exact Match Accuracy
exact_match = np.mean([inv_label_map[p] == inv_label_map[t] for p, t in zip(all_preds, all_true)])

# Compute MRR
def compute_mrr(all_logits, all_true):
    rr_total = 0.0
    for logits, true_choice in zip(all_logits, all_true):
        # logits shape (3,) - score per choice
        ranked_choices = np.argsort(logits)[::-1]
        rank = np.where(ranked_choices == true_choice)[0][0] + 1
        rr_total += 1.0 / rank
    return rr_total / len(all_true)


mrr_score = compute_mrr(all_logits, all_true)

# Compute MAP (mean average precision)
# Since this is classification, we compute AP per class and average it
true_onehot = np.eye(len(label_map))[all_true]  # shape: (num_samples, num_classes)
probs_all = np.vstack([logits.mean(axis=0) for logits in all_logits])  # mean prob over 3 choices

map_score = average_precision_score(true_onehot, probs_all, average="macro")

# Compute Macro F1
f1_macro = f1_score(all_true, all_preds, average="macro")

print(f"Exact Match Accuracy: {exact_match:.4%}")
print(f"MRR: {mrr_score:.4f}")
print(f"MAP (macro average precision): {map_score:.4f}")
print(f"Macro F1-score: {f1_macro:.4f}")
```

```python
class PubMedQAWrapper:
    def __init__(self, model, tokenizer, label_map):
        self.model = model
        self.tokenizer = tokenizer
        self.label_map = label_map
        self.inv_label_map = {v: k for k, v in label_map.items()}

    def predict_proba(self, texts):
      # Tokenize input text
      inputs = self.tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

      # Remove token_type_ids if it exists (to avoid errors)
      if "token_type_ids" in inputs:
          inputs.pop("token_type_ids")

      self.model.eval()
      with torch.no_grad():
          outputs = self.model(**inputs)
          logits = outputs.logits
          probs = torch.softmax(logits, dim=-1).cpu().numpy()
      return probs
```
