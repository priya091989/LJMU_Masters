```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = ['Baseline model','PubMedBERT','BioBERT','BioGPT', 'GPT-4']

# Replace these values with your actual results
accuracy = [0.78,0.62,0.57, 0.75, 0.60]   # Example accuracy scores
f1_score = [0.82,0.58,0.53, 0.81, 0.60]   # Example F1 scores

# Bar width and positions
x = np.arange(len(models))  # [0, 1, 2]
width = 0.35

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, accuracy, width, label='Accuracy', color='#4E79A7')
bars2 = ax.bar(x + width/2, f1_score, width, label='F1-Score', color='#F28E2B')

# Add labels, title, legend, and custom ticks
ax.set_xlabel('Models')
ax.set_ylabel('Score')
ax.set_title('Accuracy and F1-Score Comparison Across Models')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0, 1.0)
ax.legend()

# Annotate each bar with its height
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom', fontsize=9)

# Show plot
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
